<div class="alert alert-block alert-warning">
    <font size = 4><b>Note: Need For Dimensionality Reduction</b></font><br><br>
        <ul>
            <li>Dimensionality reduction is the process to reduce the number of dimensions in the feature space.</li>
            <ul>
                <li>A <b>feature = column = dimension</b>, meaning they are the same.</li>
            </ul>
            <li>In machine learning, we tend to add many features to get more accurate results. However, after a certain point, the performance and robustness of the model start decreasing and the computational complexity starts increasing as we increase the number of features. This is called <b>The Curse of Dimensionality</b> where the sample density decreases exponentially with the increase of dimensionality.</li>
            <ul>
                <li>We use dimentionality reduction to transform the data into low dimensions while keeping most of the information intact.</li>
                <li>It also helps us to visualize the high-dimensional data in 2D and 3D.</li>
                <li><b>The curse is with more features|columns|dimesion the harder it becomes to draw insights and machine learning becomes less effective.</b>
            </ul>
            <li>There are the following techniques we can use for dimensionality reduction:</li>
            <ul>
                <li><b>Principal Component Analysis (PCA)</b> - is a dimensionality reduction technique used for the identifation of a smaller number of uncorrelated variables knowns as principal compenents from a larger dataset. This technique is widely used to <b>emphasize variation and caputure strong patterns in a dataset</b>. PCA requires two properties to be true:</li>
                <ul>
                    <li>The two variables must have a linear correlation</li>
                    <li>The two variables must be on the same scale</li>
                </ul>    
                <li><b>t-distributed stochastic neighbor embedding (t-SNE)</b> is a technique for dimensionality reduction that is well suited for <b>visualization of high-dimensional datasets.</b></li>
                <ul>
                    <li>This is not a clustering technique</li>
                </ul>
            </ul>
    </ul>

<font size=3><b>Data Preprocessing</b></font>

```python
# Dropping the columns - SrNo and Date
data_air_pol.drop(columns = ["SrNo", "Date"], inplace = True)
```

```python
# Imputing missing values with mode for the Weather column and with median for all other columns
for col in data_air_pol.columns:
    
    if col == "Weather":
        data_air_pol[col].fillna(value = data_air_pol[col].mode()[0], inplace = True)
    
    else:
        data_air_pol[col].fillna(value = data_air_pol[col].median(), inplace = True)
```

```python
# Creating dummy variables for Weather column
data_air_pol = pd.get_dummies(data_air_pol, drop_first = True)
```

<font size=3><b>Scaling the data</b></font>

<div class="alert alert-block alert-warning">
    <font size = 4><b>Note:</b></font><br><br>Before starting machine learning, you have to scale the data.</div>

```python
scaler = StandardScaler()
data_air_pol_scaled = scaler.fit_transform(data_air_pol)
```

```python
data_air_pol_scaled = pd.DataFrame(data_air_pol_scaled, columns = data_air_pol.columns)
```

<font size=3><b>Principal Component Analysis (PCA)</b></font>

```python
# Defining the number of principal components to generate
n = data_air_pol_scaled.shape[1]

# Finding principal components for the data
pca1 = PCA(n_components = n, random_state = 1)
data_air_pol_pca = pd.DataFrame(pca1.fit_transform(data_air_pol_scaled))

# The percentage of variance explained by each principal component
exp_var1 = pca1.explained_variance_ratio_
```

```python
# Visualize the explained variance by individual components
plt.figure(figsize = (10, 10))

plt.plot(range(1, 29), pca1.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')

plt.title("Explained Variances by Components")

plt.xlabel("Number of Components")

plt.ylabel("Cumulative Explained Variance")

plt.show()
```

<div class="alert alert-block alert-warning">
    <font size = 4><b>Note:</b></font><br><br>The chart above is telling you that two columns is about 40% of the data. That's decent, but you want to go higher, for example 80% (.8) on the left. For example, the 5 (components) at the bottom represent just over 70% of the data. If you use 5 that means PCA reduced your 27 colums/features to 5 and it covers 70 percent of the data.</div>

```python
# Find the least number of components that can explain more than 70% variance
sum = 0

for ix, i in enumerate(exp_var1):
    
    sum = sum + i
    
    if(sum>0.70):
        print("Number of PCs that explain at least 70% variance: ", ix + 1)
        break
```

```python
# Making a new dataframe with first 5 principal components as columns and original features as indices
cols = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5']

pc1 = pd.DataFrame(np.round(pca1.components_.T[:, 0:5], 2), index = data_air_pol_scaled.columns, columns = cols)
```

<font size=3><b>Let's try to visualize the data for different perplexity values.</b></font>

```python
for i in range(10, 50, 5):
    print(i)
    tsne = TSNE(n_components = 2, random_state = 1, perplexity = i)
    
    data_air_pol_tsne = tsne.fit_transform(data_air_pol_scaled)
    
    data_air_pol_tsne = pd.DataFrame(data_air_pol_tsne)
    
    data_air_pol_tsne.columns = ['X1', 'X2']
    
    plt.figure(figsize = (7,7))
    
    sns.scatterplot(x = 'X1', y = 'X2', data = data_air_pol_tsne)
    
    plt.title("perplexity = {}".format(i))
```

<div class="alert alert-block alert-info">
    <font size=4 color=black><b>Observations:</b></font><br>
    <ul>
        <li>We can clearly see <b>4 groups in the data</b>.</li>
    </ul>
</div>

<font size=3><b>Let's label these 4 groups using the values of the X1 and X2 axes.</b></font>

```python
# Let's assign points to 4 different groups
def grouping(x):
    first_component = x['X1']
    
    second_component = x['X2']
    
    if second_component > 12:
        return 'group_1'
    
    elif (second_component < -10) and (first_component > -10): 
        return 'group_2'
    
    elif (second_component < 12) and (first_component > -9):
        return 'group_3'
    
    else: 
        return 'group_4'
```

```python
data_air_pol_tsne['groups'] = data_air_pol_tsne.apply(grouping, axis = 1)
```

```python
# Scatter plot for two components with hue
plt.figure(figsize = (7, 7))

sns.scatterplot(x = 'X1', y = 'X2', data = data_air_pol_tsne, hue = 'groups')
```
